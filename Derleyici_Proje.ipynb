```python
import sys
counter = 0
loop_counter = 0
loop_finish = 0
finish_if = 0
else_counter = 0
token = ""
language="qwertyuopasdfghjklizxcvbnm1234567890[]{}()+-/*%<>=.;?^:"
var_list = {}

def get_all_string(filename):
    file = open(filename,"r")
    text = file.read()
    text = text.replace(" ","")
    text = text.replace("\n","")
    for i in text:
        if i not in language:
            raise Exception("Token error!");
    return text

def getToken():
    text = get_all_string("string.txt")
    global counter,loop_counter,loop_finish,finish_if,else_counter
    token = text[counter]
    if token == "{":
        loop_counter = counter
        text2 = text[counter:]
        loop_finish = len(text[:counter])+text2.find("}")
    if token == "}":
        counter = loop_counter
        token = text[counter]
    if token == "[":
        text2 = text[counter:]
        finish_if = len(text[:counter])+text2.find("]")
        if text2.find(":") != -1:
            else_counter = len(text[:counter])+text2.find(":")
    if token == "]":
        counter += 1
        token = text[counter]
    if token == ":":
        counter = finish_if+1
        token = text[counter]
    if token == ".":
        print("program bitti")
        sys.exit()
    counter += 1
    return token

def P():
    token = getToken()
    C(token)
    
def C(token):
    if token == "[":
        I(token)
    elif token == "{":
        W(token)
    elif token == "<":
        Ç(token)
    elif token == ">":
        G(token)
    else:
        A(token)

def I(token):
    global counter,finish_if,else_counter
    nextToken = getToken()
    temp = E(nextToken)
    nextToken = getToken()
    if nextToken == "<":
        nextToken = getToken()
        if temp < E(nextToken):
            nextToken = getToken()
            return C(nextToken)
        else:
            if else_counter != 0:
                counter = else_counter + 1
                nextToken = getToken()
                return C(nextToken)
            counter = finish_if
            nextToken = getToken()
            return C(nextToken)
    elif nextToken == ">":
        nextToken = getToken()
        if temp > E(nextToken):
            nextToken = getToken()
            return C(nextToken)
        else:
            if else_counter != 0:
                counter = else_counter + 1
                nextToken = getToken()
                return C(nextToken)
            counter = finish_if
            nextToken = getToken()
            return C(nextToken)
    else:
        nextToken = getToken()
        if temp == E(nextToken):
            nextToken = getToken()
            return C(nextToken)
        else:
            if else_counter != 0:
                counter = else_counter + 1
                nextToken = getToken()
                return C(nextToken)
            counter = finish_if
            nextToken = getToken()
            return C(nextToken)
        
def W(token):
    global loop_finish,counter
    nextToken = getToken()
    temp = E(nextToken)
    nextToken = getToken()
    if nextToken == "<":
        nextToken = getToken()
        if temp < E(nextToken):
            nextToken = getToken()
            return C(nextToken)
        else:
            counter=loop_finish+1
            nextToken = getToken()
            return C(nextToken)
    elif nextToken == ">":
        nextToken = getToken()
        if temp > E(nextToken):
            nextToken = getToken()
            return C(nextToken)
        else:
            counter=loop_finish+1
            nextToken = getToken()
            return C(nextToken)
        
def A(token):
    global var_list
    temp = K(token,sol=True)
    nextToken = getToken()  
    if nextToken != "=":
        raise Exception("Token error!")
    nextToken = getToken()
    var_list.update({temp:E(nextToken)})
    nextToken = getToken()
    return C(nextToken)

def Ç(token):
    nextToken = getToken()
    print(E(nextToken))
    return C(getToken())

def G(token):
    global var_list
    num = str(input())
    nextToken = getToken()
    var = K(nextToken,sol=True)
    nextToken = getToken()
    if nextToken != ";":
        raise Exception("Token error!")
    var_list.update({var:num})
    return C(getToken())

def E(token):
    global var_list
    nextToken = getToken()
    if nextToken in ";?":
        return T(token)
    elif nextToken == "+":
            var1 = T(token)
            var2 = T(getToken())
            var1 = var1+var2
            return E(str(var1))
    elif nextToken == "-":
            var1 = T(token)
            var2 = T(getToken())
            var1 = var1-var2
            return E(str(var1))
    elif nextToken in "*/%^":
        return T(nextToken,token)
    else:
        raise Exception("Token error!")
        
def T(token,before_token=None):
    if before_token is not None:
        if token == "*":
            nextToken = getToken()
            var1 = T(before_token)
            var2 = T(nextToken)
            var1 = var1*var2
            return E(str(var1))
        elif token == "/":
            nextToken = getToken()
            var1 = T(before_token)
            var2 = T(nextToken)
            var1 = int(var1/var2)
            return E(str(var1))
        elif token == "%":
            nextToken = getToken()
            var1 = T(before_token)
            var2 = T(nextToken)
            var1 = int(var1%var2)
            return E(str(var1))
        elif token == "^":
            return U(token,before_token)
    if token.replace("-","").replace("+","").isnumeric() or token.isalpha():
        return U(token)
    
def U(token,before_token=None):
    if token.replace("-","").replace("+","").isnumeric() or token.isalpha():
        return F(token)
    if before_token is not None:
        if token == "^":
            nextToken = getToken()
            var1 = T(before_token)
            var2 = T(nextToken)
            var1 = int(var1**var2)
            return E(str(var1))
        
def F(token):
    if token.replace("-","").replace("+","").isnumeric():
        return R(token)
    elif token.isalpha():
        return K(token)
    
def K(token,sol=False):
    if token.isalpha():
        if sol:
            return token
        return R(str(var_list[token]))    
    raise Exception("Token error!")
    
def R(token):
    if token.replace("-","").replace("+","").isnumeric:
        return int(token)
    else:
        raise Exception("Token error!")

if __name__ == "__main__":
    P()
```
